### 09 결측값 알아보기

#### [09 - 1] 결측값이란?

```python
# numpy 라이브러리에서 결측값 불러오기
from numpy import NaN, NAN, nan
```

```python
# 결측값과의 비교
print(NaN == True)
print(NaN == 0)
print(NaN == '')
print(NaN == NaN)
print(NaN == NAN)
print(NaN == nan)
print(nan == NAN)
```

```python
import pandas as pd

# 결측값 확인
print(pd.isnull(NaN))
print(pd.isnull(nan))
print(pd.isnull(NAN))
```

```python
# 결측값이 아닌지 확인
print(pd.notnull(NaN))
print(pd.notnull(42))
print(pd.notnull('missing'))
```

##### 4. 인덱스를 다시 설정할 때 생기는 결측값

```python
gapminder = pd.read_csv('../data/gapminder.tsv', sep = '\t')

# 연도별 평균수명
life_exp = gapminder.groupby('year')['lifeExp'].mean()
print(life_exp)
```

```python
# 2000년 이후 데이터 추출
y2000 = life_exp[life_exp.index > 2000]
print(y2000)
```

```python
# 2002, 2007 제외한 나머지 인덱스 결측값 NaN 채우기
print(y2000.reindex(range(2000,2010)))
```

#### [09 - 3] 결측값 다루기

##### 1. 결측값 처리하기

##### 1) 결측값 개수 구하기

```python
# 파일 불러오기
ebola = pd.read_csv('../data/country_timeseries.csv')

# NaN이 아닌 값의 개수
print(ebola.count())
```

```python
# 전체 행 개수
num_rows = ebola.shape[0]

# 결측값 개수
num_missing = num_rows - ebola.count()
print(num_missing)
```

```python
import numpy as np

# 결측값 개수 1
print(np.count_nonzero(ebola.isnull()))     ## count_nonzero : 0(False)을 제외한 수 개수

# 특정 열의 결측값 개수
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
# 결측값 개수 2
cnts = ebola.Cases_Guinea.value_counts(dropna = False)      ## dropna = False : 비어있는 값도 포함
print(cnts)
```

```python
# 결측값 개수 3
print(cnts.loc[pd.isnull(cnts.index)])
```

```python
# 결측값 개수 4
print(ebola.Cases_Guinea.isnull().sum())
```

##### 2) 결측값 대체하기

```python
# 결측갑승ㄹ 모두 0으로 변환
print(ebola.fillna(0).iloc[:, 0:5])
```

##### 3) 정방향 채우기

```python
# 결측값 직전에 찾은 값으로 채우기 (정방향 채우기)
print(ebola.fillna(method = 'ffill').iloc[:, 0:5])      ## 열의 첫 행이 결측값이면 해당 데이터는 결측값으로 유지
```

##### 4) 역방향 채우기

```python
# 아래에서 위로 훑으면서 결측값 직전에 찾은 값으로 결측값 채우기 (역방향 채우기)
print(ebola.fillna(method = 'bfill').iloc[:, 0:5])
```

##### 5) 보간법으로 채우기

```python
# 결측값 양쪽 값의 중간값으로 채우기
print(ebola.interpolate().iloc[:, 0:5])
```

##### 6) 결측값 삭제하기

```python
# ebola의 데이터 크기 확인
print(ebola.shape)
```

```python
# 결측값이 하나도 없는 행만 남기기
ebola_dropna = ebola.dropna()
print(ebola_dropna.shape)
```

##### 7) 결측값이 있는 데이터 계산하기

```python
ebola["Cases_multiple"] = (
    ebola["Cases_Guinea"]
    + ebola["Cases_Liberia"]
    + ebola["Cases_SierraLeone"]
)
```

```python
ebola_subset = ebola.loc[:, ["Cases_Guinea", "Cases_Liberia", "Cases_SierraLeone", "Cases_multiple"]]
print(ebola_subset.head(10))
```

```python
# 결측값 건너뛰고 값 계산
print(ebola.Cases_Guinea.sum(skipna = True))

# 결측값 건너뛰지 않고 값 계산
print(ebola.Cases_Guinea.sum(skipna = False))
```

#### [09 - 4] 판다스 내장 NA 결측값 살펴보기

```python
scientists = pd.DataFrame(
    {"Name" : ["Rosaline Franklin", "William Gosset"],
     "Occupation" : ["Chemist", "Statistician"],
     "Born" : ["1920-07-25", "1876-06-13"],
     "Died" : ["1958-04-16", "1937-10-16"],
     "Age" : [37, 61]
     }
)
print(scientists)
```

```python
# 데이터형
print(scientists.dtypes)
```

```python
# 일부 값 결측값 변경
scientists.loc[1, "Name"] = pd.NA
scientists.loc[1, "Age"] = pd.NA
print(scientists)
```

```python
# 데이터형 다시 확인
print(scientists.dtypes)
```

```python
# 결측값을 np.NaN으로 설정했을 때와 비교
scientists = pd.DataFrame(
    {"Name" : ["Rosaline Franklin", "William Gosset"],
     "Occupation" : ["Chemist", "Statistician"],
     "Born" : ["1920-07-25", "1876-06-13"],
     "Died" : ["1958-04-16", "1937-10-16"],
     "Age" : [37, 61]
     }
)
print(scientists)
```

```python
# np.NaN으로 결측값으로 변환
scientists.loc[1, "Name"] = np.NaN
scientists.loc[1, "Age"] = np.NaN
print(scientists)

# 데이터형 확인
print(scientists.dtypes)      

## 결과 : np.Na를 사용하든 np.NaN을 사용하든 결과는 똑같다!
```
