### 07 데이터 결합하고 분해하기

#### [07 - 1] 데이터 연결하기

```python
import pandas as pd

df1 = pd.read_csv('../data/concat_1.csv')
df2 = pd.read_csv('../data/concat_2.csv')
df3 = pd.read_csv('../data/concat_3.csv')
```

```python
# df1, df2, df3 데이터 확인
print(df1)
print('---------------')
print(df2)
print('---------------')
print(df3)
print('---------------')
```

##### 1. 데이터프레임 살펴보기

```python
# df1 속성
print(df1.index)

# df2 열 이름
print(df1.columns)

# df1 값
print(df1.values)
```

##### 2. 행 연결하기

##### 1) 행 방향 연결하기

```python
# concat() 함수에 리스트로 전달
row_concat = pd.concat([df1, df2, df3])
print(row_concat)
```

```python
# 네 번째 행 추출
print(row_concat.iloc[3, :])
```

```python
# 시리지 하나 생성하여 데이터프레임에 연결
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])
print(new_row_series)
```

```python
# new_row_series와 df1 결합
print(pd.concat([df1, new_row_series]))
```

```python
# n1, n2, n3, n4을 데이터프레임으로 변환
new_row_df = pd.DataFrame(
    data = [["n1", "n2", "n3", "n4"]],
    columns = ["A", "B", "C", "D"],
)
print(new_row_df)
```

```python
# df1과 new_row_df 결합
print(pd.concat([df1, new_row_df]))
```

##### 2) 새로운 인덱스 설정하기

```python
row_concat_i = pd.concat([df1, df2, df3], ignore_index = True)
print(row_concat_i)
```

##### 3. 열 연결하기

##### 1) 열 방향 연결하기

```python
# df1, df2, df3 열 방향으로 연결
col_concat = pd.concat([df1, df2, df3], axis = "columns")
print(col_concat)
```

```python
# 열 이름 기준으로 데이터 추출
print(col_concat['A'])
```

```python
# 열 추가
col_concat['new_col_list'] = ['n1', 'n2', 'n3', 'n4']
print(col_concat)
```

```python
# 데이터프레임에 시리즈 추가
col_concat['new_col_series'] = pd.Series(['n1', 'n2', 'n3', 'n4'])
print(col_concat)
```

```python
# 열 이름 다시 설정
print(pd.concat([df1, df2, df3], axis = 1, ignore_index = True))
```

##### 4. 인덱스나 열 이름이 다른 데이터 연결하기

##### 1) 열 이름이 다른 데이터 행 방향 연결하기

```python
# 열 이름 변경
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
```

```python
# df1, df2, df3 데이터 확인
print(df1)
print('---------------')
print(df2)
print('---------------')
print(df3)
print('---------------')
```

```python
# df1, df2, df3 데이터 연결
row_concat = pd.concat([df1, df2, df3])
print(row_concat)
```

```python
# 공통인 열만 출력
print(pd.concat([df1, df2, df3], join = 'inner'))
```

```python
# df1, df3 공통인 열만 출력
print(pd.concat([df1, df3], join = 'inner'))
```

##### 2) 인덱스가 다른 데이터 열 방향 연결하기

```python
df1.index = [0, 1, 2, 3]
df2.index = [4, 5, 6, 7]
df3.index = [0, 2, 5, 7]
```

```python
# df1, df2, df3 데이터 확인
print(df1)
print('---------------')
print(df2)
print('---------------')
print(df3)
print('---------------')
```

```python
# 열 방향으로 연결
col_concat = pd.concat([df1, df2, df3], axis = 1)
print(col_concat)
```

```python
# df1, df3 공통인 행만 출력
print(pd.concat([df1, df3], axis = 1, join = 'inner'))
```

##### 2) 루프 구문으로 여러 개의 파일 불러오기

```python
billboard_data_files = (
    Path(".")
    .glob("../data/billboard_by_week/billboard-*.csv")
)

# 빈 리스트 생성
list_billboard_df = []

# CSV 파일명 리스트 순회
for csv_filename in billboard_data_files:
    # print(csv_filename)

    # CSV 파일을 데이터프레임으로 불러오기
    df = pd.read_csv(csv_filename)

    # 데이터프레임을 리스트에 추가
    list_billboard_df.append(df)

print(len(list_billboard_df))
```

```python
# 첫 번째 요소 유형 확인
print(type(list_billboard_df[0]))

# 데이터 확인
print(list_billboard_df[0])
```

```python
# 데이터 연결
billboard_loop_concat = pd.concat(list_billboard_df)
print(billboard_loop_concat.shape)
```

#### [07 - 4] 여러 데이터셋 병합하기

```python
person = pd.read_csv('../data/survey_person.csv')
site = pd.read_csv('../data/survey_site.csv')
survey = pd.read_csv('../data/survey_survey.csv')
visited = pd.read_csv('../data/survey_visited.csv')
```

```python
print(person)
print('---------------')
print(site)
print('---------------')
print(survey)
print('---------------')
print(visited)
```

##### 1) 일대일 병합하기

```python
# visited의 일부 데이터 추출
visited_subset = visited.iloc[[0, 2, 6], :]
print(visited_subset)
```

```python
# 중복 데이터 유무 확인
print(visited_subset["site"].value_counts())
```

```python
# site의 name열과 오른쪽 데이터프레임 visited_subset의 site 열 공통값을 기준으로 연결
o2o_merge = site.merge(visited_subset, left_on = "name", right_on = "site")
print(o2o_merge)
```

##### 2) 다대일 병합하기

```python
# 중복값 개수 확인
print(visited["site"].value_counts())
```

```python
m2o_merge = site.merge(visited, left_on = 'name', right_on = 'site')
print(m2o_merge)
```

##### 3) 다대다 병합하기

```python
ps = person.merge(survey, left_on = 'ident', right_on = 'person')
vs = visited.merge(survey, left_on = 'ident', right_on = 'taken')
```

```python
print(ps)
print('-----------------------')
print(vs)
```

```python
# quant 열 중복값 확인
print(ps["quant"].value_counts())
print('---------------')
print(vs["quant"].value_counts())
```

```python
# quant 기준으로 병합
ps_vs = ps.merge(vs, left_on = "quant", right_on = "quant")
print(ps_vs.loc[0, :])
```

#### [07 - 5] 데이터 정규화하기

##### 1) 표 분할하여 데이터 정규화하기

```python
billboard = pd.read_csv('../data/billboard.csv')

# 피벗 되돌리기
billboard_long = billboard.melt(
    id_vars = ["year", "artist", "track", "time", "date.entered"],
    var_name = "week",
    value_name = "rating",
)
print(billboard_long)
```

```python
# 특정 곡 데이터 추출
print(billboard_long.loc[billboard_long.track == 'Loser'])
```

```python
# 새로운 데이터프레임으로 옮길 4개의 열 데이터 추출
billboard_songs = billboard_long[["year", "artist", "track", "time", "date.entered"]]
print(billboard_songs.shape)
```

```python
# 중복된 값 제거
billboard_songs = billboard_songs.drop_duplicates()
print(billboard_songs.shape)
```

```python
# 인덱스 값에 1을 더한 id 할당
billboard_songs['id'] = billboard_songs.index + 1
print(billboard_songs)
```

```python
# 곡 정보와 관련된 4개 열 기준으로 데이터프레임 병합
billboard_ratings = billboard_long.merge(
    billboard_songs, on = ["year", "artist", "track", "time", "date.entered"]
)
print(billboard_ratings.shape)
```

```python
# 결과 확인
print(billboard_ratings)
```

```python
# 주별 순위 데이터프레임
billboard_ratings = billboard_ratings[["id", "week", "rating"]]
print(billboard_ratings)
```
