##### 1) 문자 추출하기

```python
word = 'grail'
sent = 'a scratch'
```

##### 2) 음수 인덱스로 추출하기

```python
print(sent[-1])
print(sent[-9:-8])
print(sent[0:-8])
print(sent[2:-1])   ## 음수 인덱스로 추출해도 오른쪼고 인덱스는 추출 범위에 포함하지 않음
print(sent[-7:-1])
```

```python
s_len = len(sent)
print(s_len)

print(sent[2:s_len])
```

##### 4) 슬라이싱 구문에서 왼쪽이나 오른쪽 인덱스를 지정하지 않고 문자 추출하기

```python
print(word[0:3])
# 왼쪽 인덱스 생략
print(word[ : 3])

print(sent[2:len(sent)])
# 오른쪽 인덱스 생략
print(sent[2:])

# 양쪽 인덱스 생략
print(sent[:])
```

##### 5) 슬라이싱 간격 설정하기

```python
# 2개 간격으로 문자 추출
print(sent[::2])

# 3개 간격으로 문자 추출
print(sent[::3])
```

14. split() : 구분 문자를 기준으로 문자열을 나눈 리스트를 반환
15. partition() : split() 메서드와 비슷한 역할을 하지만, 구분 기호도 함께 반환
16. center() : 지정한 너비로 문자열을 늘이고 가운데 정렬
17. zfill() : 문자열 빈칸을 '0'으로 채움

#### [11 - 3] 문자열 메서드 더 알아보기

##### 1) join() 메서드

```python
# 문자 연결
d1 = '40°'
m1 = "46'"
s1 = '52.837"'
u1 = 'N'

d2 = '73°'
m2 = "58'"
s2 = '26.302"'
u2 = 'W'

coords = ' '.join([d1, m1, s1, u1, d2, m2, s2, u2])
print(coords)
```

##### 2) splitlines() 메서드

```python
multi_str = """Guard : What? Ridden on a horse?
King Arthur : Yes!
Guard : You're using coconuts!
King Arthur : What?
Guard : You've god ... coconut[s] and you're bangin' 'em together.
"""
print(multi_str)
```

```python
# 각 줄을 나누어 요소로 저장
multi_str_split = multi_str.splitlines()
print(multi_str_split)
```

```python
# Guard:로 시작하는 대화 문장만 추출
guard = multi_str_split[::2]
print(guard)
```

```python
# 인물 정보 제외
guard = multi_str.replace('Guard : ', '').splitlines()[::2]
print(guard)
```

##### [11 - 5] 정규식으로 문자열 처리에 날개 달기

[정규식 문법]
1. . : 임의의 문자 1개를 나타낸다.
2. ^ : 문자열 처음부터 일치하는 패턴을 찾는다.
3. $ : 문자열 긑부터 일치하는 패턴을 찾는다.
4. \* : *앞의 문자가 0개 이상인 패턴을 찾는다.
5. \+ : +앞의 문자가 1개 이상인 패턴을 찾는다.
6. ? : ?앞의 문자가 있거나 없는 패턴을 찾는다.
7. {m} : 앞의 문자가 m개 반복되는 패턴을 찾는다.
8. {m, n} : 앞의 문자가 m개 이상, n개 이하로 반복되는 패턴을 찾는다.
9. \ : *, ?, +와 같은 특수 문자를 검색할 때 이스케이프 문자(\)를 사용한다.
10. [] : [] 사이의 문자 중 하나와 일치하는 패턴을 찾는다. 예를 들어 [a-z]는 a부터 z까지 범위의 알파벳 중 하나와 일치하는 패턴을 찾는다.
11. | : OR 연산을 나타낸다. A | B는 A또는 B와 일치하는 패턴을 찾는다.
12. () : ()에 지정한 패턴을 찾을 때 사용한다.

[특수 문자]
1. \d : 숫자 1개를 의미한다([0-9]와 같음).
2. \D : 숫자 이외의 문자 1개를 의미한다.([^[0-9]와 같음).
3. \s : 공백이나 탭 등의 공백 문자 1개를 의미한다.
4. \S : \s에 해당하지 않는 문자 1개를 의미한다.
5. \w : 알파벳 또는 숫자, 밑줄 1개를 의미한다.
6. \W : \w에 해당하지 않는 문자 1개를 의미한다.

[정규식 메서드]
1. search() : 첫 번째로 찾은 패턴의 양쪽 끝 인덱스를 반환한다.
2. match() : 문자열의 처음부터 검색하여 찾아낸 패턴의 양쪽 끝 인덱스를 반환한다.
3. fullmatch() : 전체 문자열이 일치하는지 검사한다.
4. split() : 지정한 패턴으로 나눈 문자열을 리스트로 반환한다.
5. findall() : 지정한 패턴을 찾아 리스트로 반환한다.
6. finditer() : findall() 메서드와 기능이 같지만, 이터레이터를 반환한다.
7. sub() : 첫 번째 인자로 전달한 패턴을 두 번째 인자로 전달한 문자열로 교체한다.

##### 1) 정규식으로 패턴 찾기

```python
import re

tele_num = '1234567890'
```

```python
# 연속된 숫자 10개 \d 기호 사용하여 찾기
m = re.match(pattern = '\d\d\d\d\d\d\d\d\d\d', string = tele_num)
print(m)
print(type(m))
```

```python
# 패턴과 일치하는 문자열이 있는지 True 또는 False라는 값으로 확인
print(bool(m))

## 조건문 사용
if m:
    print('match')
else:
    print('no match')
```

```python
# start(), end(), span(), group()
print(m.start())
print(m.end())
print(m.span())
print(m.group())
```

```python
tele_num_spaces = '123 456 7890'

m = re.match(pattern = '\d{10}', string = tele_num_spaces)
print(m)

if m:
    print('match')
else:
    print('no match')
```

```python
# 3개의 숫자, 공백, 3개의 숫자, 공백, 마지막 4개의 숫자가 이어지는 패턴
p = '\d{3}\s?\d{3}\s?\d{4}'
m = re.match(pattern = p, string = tele_num_spaces)
print(m)
```

```python
tele_num_space_paren_dash = '(123) 456-7890'
p = '\(?\d{3}\)?\s?\d{3}\s?-?\d{4}'
m = re.match(pattern = p, string = tele_num_space_paren_dash)
print(m)
```

```python
city_tele_num_space_paren_dash = '+1 (123) 456-7890'
p = '\+1\s\(?\d{3}\)?\s?\d{3}\s?-?\d{4}'
m = re.match(pattern = p, string = city_tele_num_space_paren_dash)
print(m)
```

##### 2) 알기 쉬운 정규식 만들기

```python
p = (
    '\+?'       ## +가 0개 또는 1개
    '1'         ## 숫자 1
    '\s?'       ## 공백 문자가 0개 또는 1개
    '\(?'       ## ( 문자가 0개 또는 1개
    '\d{3}'     ## 숫자 3개
    '\)?'       ## ) 문자가 0개 또는 1개
    '\s?'       ## 공백 문자가 0개 또는 1개
    '\d{3}'     ## 숫자 3개
    '\s?'       ## 공백 문자가 0개 또는 1개
    '-?'        ## -가 0개 또는 1개
    '\d{4}'     ## 숫자 4개
)
print(p)
```

```python
cnty_tele_num_space_paren_dash = '+1 (123) 456-7890'
m = re.match(pattern = p, string = cnty_tele_num_space_paren_dash)
print(m)
```

##### 3) 패턴과 일치하는 모든 문자열 찾기

```python
s = (
    "14 Ncuti Gatwa,"
    "13 Jodie Whittaker, war John Hurt, 12 Peter Capaldi, "
    "11 Matt Smith, 10 David Tennant, 9 Christopher Eccleston"
)
print(s)
```

```python
# 문자열에서 숫자가 1개 이상 연속된 모든 부분 찾기
p = "\d+"
m = re.findall(pattern = p, string = s)
print(m)
```

##### 4) 패턴과 일치하는 문자열 대체하기

```python
multi_str = """Guard: What? Ridden on a horse?
King Arthur: Yes!
Guard: You're using coconuts!
King Arthur: What?
Guard: You've god ... coconut[s] and you're bangin' 'em together.
"""
```

```python
# 모든 인물 정보 제거
p = '\w+\s?\w+:\s?'
s = re.sub(pattern = p, string = multi_str, repl = '')
print(s)
```

```python
# Guard와 King Arthur가 한 말 각각 추출
guard = s.splitlines()[::2]
kinga = s.splitlines()[1::2]
print("Guard가 한 말 => ", guard)
print("Kinga가 한 말 => ", kinga)
```

##### 5) compile() 함수

```python
p = re.compile('\d{10}')
s = '1234567890'
m = p.match(s)
print(m)
```

```python
p = re.compile('\d+')
s = (
    "14 Ncuti Gatwa, " 
    "13 Jodie Whittaker, war John Hurt, 12 Peter Capaldi, "
    "11 Matt Smith, 10 David Tennant, 9 Christopher Eccleston"
)
m = p.findall(s)
print(m)
```

```python
p = re.compile('\w+\s?\w+:\s?')
s = "Guard: You're using coconuts!"
m = p.sub(string = s, repl = '')
print(m)
```
