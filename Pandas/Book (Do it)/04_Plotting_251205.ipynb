#### [04 - 1] 데이터 시각화란?

```python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
print(anscombe)
```

#### [04 - 2] matplotlib 라이브러리란?

```python
import matplotlib.pyplot as plt
```

```python
# Ⅰ 데이터 그래프 작성 (선 그래프)
dataset_1 = anscombe[anscombe['dataset'] == 'I']
plt.plot(dataset_1['x'], dataset_1['y'])
plt.show()
```

```python
# 점 그래프 (점 그래프)
plt.plot(dataset_1['x'], dataset_1['y'], 'o')
plt.show()
```

```python
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

##### 1) 한 번에 4개의 그래프 그리기

```python
# 그림 영역 Figure 객체인 fig 생성
fig = plt.figure()

# 빈 하위 그래프 그림 영역에 삽입
axes1 = fig.add_subplot(2, 2, 1)    # 2행 2열 중 1번 위치
axes2 = fig.add_subplot(2, 2, 2)    # 2행 2열 중 2번 위치
axes3 = fig.add_subplot(2, 2, 3)    # 2행 2열 중 3번 위치
axes4 = fig.add_subplot(2, 2, 4)    # 2행 2열 중 4번 위치

# 그래프 값 채우기
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

# 하위 그래프 제목 삽입
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

# 그림 영역 제목 삽입
fig.suptitle("Anscombe Data")

# 하위 그래프 사이 제목 삽입 공간 확보
fig.set_tight_layout(True)

plt.show()
```

#### [04 - 3] matplotlib으로 그래프 그리기

```python
tips = sns.load_dataset("tips")
print(tips)
```

##### 1. 일변량 그래프

##### 1) 히스토그램

```python
# 맑은 고딕 글꼴 설정
plt.rc('font', family = 'Malgun Gothic')

# 그림 영역 객체 fig 생성
fig = plt.figure()

# 1행, 1열, 1번 위치에 하위 그래프 삽입
axes1 = fig.add_subplot(1, 1, 1)

# 히스토그램 생성
axes1.hist(tips['total_bill'], bins = 10)

# 제목과 레이블 삽입
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Frequency')

plt.show()
```

##### 2. 이변량 그래프

##### 1) 산점도 그래프

```python
# 그림 영역 객체 생성
scatter_plot = plt.figure()

# 1행 1열 중 1번째 위치에 하위 그래프 생성
axes1 = scatter_plot.add_subplot(1, 1, 1)

# 산점도(scatter plot) 생성
# x축 : total_bill, y축 : tip
axes1.scatter(tips['total_bill'], tips['tip'])

# 그래프 제목 설정
axes1.set_title('Scatterplot of Total Bill vs Tip')

# x축 레이블 설정
axes1.set_xlabel('Total Bill')

# y축 레이블 설정
axes1.set_ylabel('Tip')

# 그래프 화면에 표시
plt.show()
```

##### 2) 박스 그래프

```python
# 그림 영역 객체 생성
boxplot = plt.figure()

# 1행 1열 중 1번째 위치에 하위 그래프 생성
axes1 = boxplot.add_subplot(1, 1, 1)

# 박스 그래프(boxplot) 생성
axes1.boxplot(
    x = [
        tips[tips['sex'] == 'Female']['tip'],
        tips[tips['sex'] == 'Male']['tip']
    ],
    labels = ['Female', 'Male']
)

# x축 레이블 생성
axes1.set_xlabel('Sex')

# y축 레이블 생성
axes1.set_ylabel('Tip')

# 그래프 제목 설정
axes1.set_title('Boxplot of Tips by Sex')

# 그래프 화면에 표시
plt.show()
```

##### 3. 다변량 그래프

##### 1) 변수가 여러 개인 그래프 그리기

```python
# 성별에 따라 점 색상을 지정하기 위한 색상 매핑 딕셔너리
colors = {"Female" : "#f1a340", "Male" : "#998ec3"}

# 그림 영역 객체 생성
scatter_plot = plt.figure()

# 1행 1열 중 첫 번째 subplot 생성
axes1 = scatter_plot.add_subplot(1, 1, 1)

# 산점도(scatter plot) 생성
axes1.scatter(data = tips,
              x = 'total_bill',
              y = 'tip',
              s = tips['size']**2*10,
              c = tips['sex'].map(colors),
              alpha = 0.5)

# 개별 Axes 제목 삽입
axes1.set_title('Colored by Sex and Sized by Size')

# x축, y축 레이블 설정
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

# 그림 전체 제목 삽입
scatter_plot.suptitle('Total Bill vs Tip')

# 그래프 화면에 표시
plt.show()
```

##### [04 - 4] seaborn으로 그래프 그리기

```python
sns.set_context("paper")
```

##### 1. 다양한 그래프 그려 보기

##### 1) 일변량 그래프 그리기

```python
# 히스토그램

## 그림 전체와 하위 그래프 영역 한 번에 생성                                                                  
hist, ax = plt.subplots()

## 히스토그램 생성
sns.histplot(data = tips, x = 'total_bill', ax = ax)    # ax라는 그래프 영역에 히스토그램 생성

# 제목 지정
ax.set_title('Total Bill HListogram')

# 그래프 화면에 표시
plt.show()
```

```python
# 밀도 분포 그래프

## Figure와 Axes 두 개의 객체 반환
den, ax = plt.subplots()

## 커널 밀도 추정 KDE 그래프 생성
sns.kdeplot(data = tips, x = 'total_bill', ax = ax)

## 제목 지정
ax.set_title('Total Bill Density')

## x축 레이블 생성
ax.set_xlabel('Total Bill')

## y축 레이블 생성
ax.set_ylabel('Unit Probability')

## 그래프 화면에 표시
plt.show()
```

```python
# 러그 그래프

## 그림 영역 객체와 하위 그래프 객체 생성
rug, ax = plt.subplots()

## 러그플롯(rug plot) 생성
sns.rugplot(data = tips, x = 'total_bill', ax = ax)

## 히스토그램 생성
sns.histplot(data = tips, x = 'total_bill', ax = ax)

## 제목 삽입
ax.set_title('Rug Plot and Histogram of Total Bill')

## 그래프 화면에 출력
plt.show()
```

```python
# 분포 그래프

## 히스토그램 + KDE + Rug plot을 하나의 그림으로 만듦
fig = sns.displot(data = tips, x = 'total_bill', kde = True, rug = True)

## 축 레이블 설정
fig.set_axis_labels(x_var = 'Total Bill', y_var = 'Count')

## 제목 설정
fig.figure.suptitle('Distrilbution of Total Bill')

## 그래프 화면에 표시
plt.show()
```

```python
# 막대그래프

## 그림 전체와 그래프 영역 객체 생성
count, ax = plt.subplots()

## 막대그래프(countplot) 생성
sns.countplot(data = tips, x = 'day', palette = 'viridis', ax = ax)

## 그래프 제목 설정
ax.set_title('Count of days')

## x축 레이블 설정
ax.set_xlabel('Day of the Week')

## y축 레이블 설정
ax.set_ylabel('Frequency')

## 그래프 화면에 표시
plt.show()
```

##### 2) 이변량 그래프

```python
# 산점도 그래프 1

## Figure와 Axes 객체 동시에 생성
scatter, ax = plt.subplots()

## 산점도 그래프 생성
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', ax = ax)

# 제목 삽입
ax.set_title('Scatter Plot of Total Bill and Tip')

# x축 레이블 설정
ax.set_xlabel('Total Bill')

# y축 레이블 설정
ax.set_ylabel('Tip')

# 그래프 화면에 표시
plt.show()
```

```python
# 산점도 그래프 그리기 2

## Figure 객체, Axes 객체 생성
reg, ax = plt.subplots()

## 회귀 그래프(regression plot) 생성
sns.regplot(data = tips, x = 'total_bill', y = 'tip', ax = ax)      # 회귀선 없이 그래프 작성 : fit_reg = False 추가

## 그래프 제목 설정
ax.set_title('Regression Plot of Total Bill and Tip')

## x축 레이블 설정
ax.set_xlabel('Total Bill')

## y축 레이블 설정
ax.set_ylabel('Tip')

## 그래프 화면에 표시
plt.show()
```

```python
## sns.lmplot()으로 산점도 그래프 그리기
fig = sns.lmplot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
# 조인트 그래프

## 산점도 그래프와 일변량 그래프 작성
joint = sns.jointplot(data = tips, x = 'total_bill', y = 'tip')

## scatter plot과 히스토그램의 축 레이블 설정
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')

## Figure 전체 제목 설정
joint.figure.suptitle('Joint Plot of Total Bill and Tip', y = 1.03)

## 그래프 화면에 표시
plt.show()
```

```python
# 육각 그래프

## 육각 그래프 그리기
hexbin = sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')

## x축, y축 레이블 설정
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')

## 그래프 제목 삽입
hexbin.figure.suptitle('Hexbin Joint Plot of Total Bill and Tip', y = 1.03)

## 그래프 화면에 표시
plt.show()
```

```python
# 2차원 밀도 분포 그래프 1

## 그래프 공간 생성
kde, ax = plt.subplots()

## 2차원 밀도 분포 그래프 생성
sns.kdeplot(data = tips, x = "total_bill", y = "tip", fill = True, ax = ax)     # fill = True : 음영 효과

## 그래프 제목 삽입
ax.set_title('Kernel Density Plot of Total Bill and Tip')

## x축, y축 레이블 삽입
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

## 그래프 화면에 표시
plt.show()
```

```python
# 2차원 밀도 분포 그래프 2

## 밀도 분포 그래프 삽입
kde2d = sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "kde")

## x축, y축 레이블 삽입
kde2d.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')

## 그래프 제목 삽입
kde2d.figure.suptitle('Hexbin Joint Plot of Total Bill and Tip', y = 1.03)

## 그래프 화면에 표시
plt.show()
```

```python
# 막대그래프

import numpy as np

## 그릠 영역과 하위 그래프 객체 생성
bar, ax = plt.subplots()

## 막대그래프 생성
sns.barplot(data = tips, x = "time", y = "total_bill", estimator = np.mean, ax = ax)

## 그래프 제목 삽입
ax.set_title('Bar Plot of Average Total Bill for Time of Day')

## x축, y축 레이블 생성
ax.set_xlabel('Time of Day')
ax.set_ylabel('Average Total Bill')

## 화면에 그래프 표시
plt.show()
```

```python
# 박스 그래프

## 그림 영역과 하위 그래프 객체 생성
box, ax = plt.subplots()

## 박스 그래프 삽입
sns.boxplot(data = tips, x = 'time', y = 'total_bill', ax = ax)

## 그래프 제목 삽입
ax.set_title('Bar Plot of Total Bill for Time of Day')

## x축, y축 레이블 삽입
ax.set_xlabel('Time of Day')
ax.set_ylabel('Total Bill')

## 화면에 그래프 표시
plt.show()
```

```python
# 관계 그래프 그리기 1

## 관계 그래프 생성
fig = sns.pairplot(data = tips)

## 제목 설정
fig.figure.suptitle('Pairwise Relationships of the Tips Data', y = 1.03)    # 제목이 그래프와 겹치지 않도록 1.03 만큼 올림

## 그래프 출력
plt.show()
```

```python
# 관계 그래프 그리기 2

## PairGrid 생성
pair_grid = sns.PairGrid(tips, diag_sharey = False)

## 대각선 위 회귀선 포함 산점도 그래프
pair_grid = pair_grid.map_upper(sns.regplot)

## 대각선 아래 이차원 밀집분포 그래프
pair_grid = pair_grid.map_lower(sns.kdeplot)

## 대각선 히스토그램
pair_grid = pair_grid.map_diag(sns.histplot)

## 그래프 출력
plt.show()
```

```python
# 바이올린 그래프

## Figure와 Axes 생성
violin, ax = plt.subplots()

## 바이올린 그래프 생성
sns.violinplot(data = tips,
               x = "time",
               y = "total_bill",
               hue = "smoker",
               split = True,
               palette = "viridis",
               ax = ax
               )

## 그래프 출력
plt.show()
```
