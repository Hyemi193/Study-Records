### 05 깔끔한 데이터 만들기

#### [05 - 1] 깔끔한 데이터란?

1. 변수는 열로 나타내야 한다.

#### [05 -2] 열 이름이 값일 때

##### 1. 하나의 열만 남기기

##### 1) 넓은 데이터 확인하기

```python
import pandas as pd

pew = pd.read_csv('../data/pew.csv')

print(pew.iloc[:, 0:5])
```

##### 2) 긴 데이터로 만들기

```python
# religion열 제외한 나머지 모든 열 피벗 되돌리기
pew_long = pew.melt(id_vars = 'religion')
print(pew_long)
```

```python
# var_name 으로 "income"을, value_name으로 "count"를 설정
pew_long = pew.melt(id_vars = "religion", var_name = "income", value_name = "count")
print(pew_long)
```

##### 2. 여러 개의 열 남기기

##### 1) 여러 개 열 유지하기

```python
billboard = pd.read_csv('../data/billboard.csv')
print(billboard.iloc[:, 0:16])
```

```python
# year, artist, track, time, date.entered 5개 열 그대로, 주와 관련된 열만 피벗 되돌리기
billboard_long = billboard.melt(
    id_vars = ["year", "artist", "track", "time", "date.entered"],
    var_name = "week",
    value_name = "ragting"
)
print(billboard_long)
```

#### [05 - 3] 열 이름에 변수가 여러 개일 때

##### 1. 열 이름이 여러 가지 뜻일 때

##### 1) 깔끔한 데이터 만들기 1

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
print(ebola.columns)
```

```python
# 몇 개의 행과 열 살펴보기
print(ebola.iloc[:5, [0, 1, 2, 10]])
```

```python
# 피벗 되돌리기
ebola_long = ebola.melt(id_vars = ['Date', 'Day'])
print(ebola_long)
```

##### 2) 깔끔한 데이터 만들기 2

```python
variable_split = ebola_long.variable.str.split('_')
print(variable_split)
```

```python
# 값 유형 확인
print(type(variable_split))

# 시리즈 첫 번째 요소 type()
print(type(variable_split[0]))
```

```python
# 리스트의 각 요소를 사례(status_values)와 국가(country_values)로 저장하기
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)
```

```python
# 잘 나뉘어졌는지 확인
print(status_values)
```

```python
# 데이터 프레임에 추가
ebola_long['status'] = status_values
ebola_long['country'] = country_values
print(ebola_long)
```

##### 2. 한 번에 분할하고 합치기

```python
# 피벗 되돌리기 상태로 초기화
ebola_long = ebola.melt(id_vars = ['Date', 'Day'])
```

```python
variable_split = ebola_long.variable.str.split('_', expand = True)  ## expand : 분할과 동시에 데이터 프레임 생성
print(variable_split)
```

```python
# 새로 분할한 열을 원본 데이터프레임에 바로 할당
ebola_long[['status', 'country']] = variable_split
print(ebola_long)
```

#### [05 - 4] 변수가 행과 열 모두에 있을 때

##### 1) 행과 열 모두에 있는 변수 정리하기

```python
# weather.csv 데이터셋 불러오기
weather = pd.read_csv('../data/weather.csv')
print(weather.iloc[:5, :11])
```

```python
# 피벗 되돌리기
weather_melt = weather.melt(
    id_vars = ["id", "year", "month", "element"],
    var_name = "day",
    value_name = "temp",
)
print(weather_melt)
```

```python
# 피벗
weather_tidy = weather_melt.pivot_table(
    index = ['id', 'year', 'month', 'day'],
    columns = 'element',
    values = 'temp'
)
print(weather_tidy)
```

```python
# 평면화
weather_tidy_flat = weather_tidy.reset_index()
print(weather_tidy_flat)
```

```python
# 메서드 체인을 이용하여 한 번에 수행하기
weather_tidy = (
    weather_melt.pivot_table(
        index = ['id', 'year', 'month', 'day'],
        columns = 'element',
        values = 'temp')
        .reset_index()
    )
print(weather_tidy)
```
