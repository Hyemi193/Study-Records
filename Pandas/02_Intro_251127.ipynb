#### [02 - 4] 기초 통계 계산하기

##### 1. 그룹화한 데이터의 평균 구하기

##### 1) 1개 열 그룹화하기

```python
# 연도별 평균 기대 수명
print(df.groupby('year')['lifeExp'].mean())
```

```python
# 단계별로 확인
## 1. 그룹화된 객체 생성
grouped_year_df = df.groupby('year')
print('그룹화 => ', grouped_year_df)
print(type(grouped_year_df))

## 2. 기대 수명을 나타내는 lifeExp 열 선택
grouped_year_df_lifeExp = grouped_year_df['lifeExp']
print("기대수명 => ", grouped_year_df_lifeExp)
print(type(grouped_year_df_lifeExp))

## 3. 평균
mean_lifeExp_by_year = grouped_year_df_lifeExp.mean()
print("평균 => ", mean_lifeExp_by_year)
```

##### 2) 2개 이상 열 그룹화하기

```python
# 리스트 사용하여 그룹화할 열 지정
multi_group_var = df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
print(multi_group_var.head(10))
```

```python
# 데이터프레임 평탄화 : reset_index()
flat = multi_group_var.reset_index()
print(flat.head(10))
```

##### 2. 그룹화한 데이터 개수 세기

##### 1) 그룹화한 데이터 개수 구하기

```python
## continent별 country 개수 : nunique()
print(df.groupby('continent')['country'].nunique())
```

```python
# 열이나 행의 개수(빈도 수) : value_counts()
print(df.groupby('continent')['country'].value_counts())
```
