#### [12 - 6] datetime 객체의 메서드 활용하기

```python
banks = pd.read_csv('../data/banklist.csv')
print(banks.head())
print(banks.info())
```

```python
# 데이터 불러올 때 두 열을 datetime형으로 불러오기
banks = pd.read_csv('../data/banklist.csv', parse_dates = ["Closing Date", "Updated Date"])
print(banks.info())
```

```python
# Closing Date 열의 분기와 연도 저장
banks = banks.assign(closing_quarter = banks['Closing Date'].dt.quarter,
                     closing_year = banks['Closing Date'].dt.year)
```

```python
# closing_year 열을 기준으로 데이터셋을 그룹화
closing_year = banks.groupby(['closing_year']).size()
print(closing_year)
```

```python
# closing_year과 closing_quarter 기준으로 그룹화
closing_year_q = (
    banks.groupby(['closing_year', 'closing_quarter']).size()
)
print(closing_year_q)
```

```python
import matplotlib.pyplot as plt

# 연도별, 분기별 파산 은행 개수
fig, ax = plt.subplots()
ax = closing_year.plot()
plt.show()
```

```python
fig, ax = plt.subplots()
ax = closing_year_q.plot()
plt.show()
```

```python
# 에볼라 데이터셋 다시 불러와서 초기화
ebola = pd.read_csv("../data/country_timeseries.csv", parse_dates = ["Date"], index_col = "Date")
print(ebola.iloc[:, :4])
```

```python
# 시간 범위 생성
new_idx = pd.date_range(ebola.index.min(), ebola.index.max())
print(new_idx)      ## 가장 오래된 날짜부터 오름차순으로 정렬 됨 
```

```python
# 오름차순으로 뒤집기
new_idx = reversed(new_idx)
print(new_idx)
```

```python
# 인덱스 다시 설정
ebola = ebola.reindex(new_idx)
print(new_idx)
```

```python
print(ebola.iloc[:, :4])
```

```python
# 각 열의 NaN이 아닌 가장 오래된 유효 날짜
last_valid = ebola.apply(pd.Series.last_valid_index)
print(last_valid)
```

```python
# 가장 오래된 날짜
earliest_date = ebola.index.min()
print(earliest_date)
```

```python
# 각 열의 가장 오래된 유효 날짜에서 earliest_date 빼기
shift_values = last_valid - earliest_date
print(shift_values)
```

```python
# 빈 딕셔너리 생성
ebola_dict = {}

# 반복문 시작 : 나라 하나씩 꺼내기
for idx, col in enumerate(ebola):
    ## 며칠이나 옮겨야 하는지 확인
    d = shift_values[idx].days
    ## 실제로 데이터 이동시키기
    shifted = ebola[col].shift(d)
    ## 옮겨진 데이터를 딕셔너리에 저장
    ebola_dict[col] = shifted
```

```python
# 데이터프레임으로 변환
ebola_shift = pd.DataFrame(ebola_dict)
```

```python
# 데이터 확인
print(ebola_shift.tail())
```

```python
# 인덱스를 제거하고 발병일을 기준을 며칠이 지났는지를 나타내는 Day 열을 인덱스로 다시 지정
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Day'], axis = "columns")
print(ebola_shift.tail())
```

```python
# 뉴욕 기준 출발시간
depart = pd.Timestamp('2017-08-29 07:00', tz = 'US/Eastern')
print(depart)
```

```python
arrive = pd.Timestamp('2017-08-29 09:57')   ## 어디 시간인지 아직 모름
print(arrive)
```

```python
# 위치 지정
arrive = arrive.tz_localize('US/Pacific')   
print(arrive)
```

```python
# 뉴욕 기준 시간
print(arrive.tz_convert('US/Eastern'))
```

```python
# 비행 시간
duration = arrive - depart  
print(duration)
```
