### 10 자료형 더 알아보기

#### [10 - 1] 자료형 살펴보기

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
# 데이터 자료형 확인
print(tips.dtypes)
```

```python
# 데이터 확인
print(tips.head())
```

#### [10 - 2] 자료형 변환하기

##### 1) 문자열로 변환하기

```python
# sex열을 object형으로 변환
tips['sex_str'] = tips['sex'].astype('str')
print(tips.dtypes) 
```

##### 2) 숫자로 변환하기

```python
# total_bill 열을 object형으로 변환
tips['total_bill'] = tips['total_bill'].astype('str')
print(tips.dtypes)
```

```python
# 다시 total_bill의 object형을 float64형으로 변환
tips['total_bill'] = tips['total_bill'].astype('float')
print(tips.dtypes)
```

##### 3) 숫자형으로 변환하는 to_numeric() 함수 사용하기

```python
# 데이터셋의 첫 10개 행을 추출하여 저장하고 total_bill 열의 몇개 값을 'missing'으로 변환
tips_sub_miss = tips.head(10).copy()
tips_sub_miss.loc[[1, 3, 5, 7], 'total_bill'] = 'missing'

print(tips_sub_miss.head(10))
```

```python
# total_bill 열 자료형 확인
print(tips_sub_miss.dtypes)
```

```python
# total_bill 열을 다시 float64형으로 변환
# pd.to_numeric(tips_sub_miss['total_bill'])    ## 'missing'이라는 문자열을 부동 소수점으로 변환하는 방법을 모르기 때문
```

to_numeric() 함수의 매개변수 errors에 설정할 수 있는 값
1. 'raise' : 변환할 수 없는 값이 있을 때 오류를 발생시킨다.
2. 'coerce' : 변환할 수 없는 값이 있을 때 NaN으로 값을 설정한다.
3. 'ignore' : 변환할 수 없는 값이 있을 때 해당 값을 그대로 사용한다.

```python
# ignore 사용
tips_sub_miss['total_bill'] = pd.to_numeric(tips_sub_miss['total_bill'], errors = 'ignore')

print(tips_sub_miss.head(10))
```

```python
# 자료형 확인
print(tips_sub_miss.dtypes)     ## 자료형 변하지 않음
```

```python
# coerce 사용
tips_sub_miss['total_bill'] = pd.to_numeric(tips_sub_miss['total_bill'], errors = 'coerce')
print(tips_sub_miss)
```

```python
# 자료형 확인
print(tips_sub_miss.dtypes)     ## float64로 변환 성공
```

#### [10 - 3] 범주형 데이터 알아보기

##### 1) 범주형으로 변환하기

```python
# sex 열을 문자열로 변환 후 메모리 사용량 확인
tips['sex'] = tips['sex'].astype('str')
print(tips.info())      ## 10.8+ KB
```

```python
# 다시 category형으로 변환 후 메모리 사용량 확인
tips['sex'] = tips['sex'].astype('category')
print(tips.info())      ## 9.3+ KB
```
